# Измерение магнитного поля Земли с помощью неодимовых магнитов

## Аннотация

В данной лабораторной работе исследуются физические свойства постоянных неодимовых магнитов. С их помощью производится измерение горизонтальной и вертикальной составляющих вектора индукции магнитного поля Земли, а также определяется магнитное наклонение в месте проведения эксперимента.

## Теоретическая справка

### Магнитный диполь

Простейшим источником магнитного поля, или **магнитным диполем**, может служить виток с электрическим током или небольшой постоянный магнит. Основной характеристикой диполя является его **магнитный момент** $\vec{\mathfrak{m}}$. Для плоского витка с током $I$ и площадью $S$ он определяется как:

$$
\vec{\mathfrak{m}}=\frac{I}{c}\,\vec{S}
$$

где $\vec{S}$ — вектор площади, перпендикулярный плоскости витка, а его направление связано с направлением тока правилом правого винта. В системе СИ коэффициент $1/c$ отсутствует.

Если размеры источника поля малы по сравнению с расстоянием до точки наблюдения, его можно считать **точечным диполем**. Поле такого диполя в вакууме описывается формулой, аналогичной полю электрического диполя:

$$
\vec{B}(\vec{r})=\frac{3\bigl(\vec{\mathfrak{m}}\cdot\vec{r}\bigr)\vec{r}}{r^{5}}-\frac{\vec{\mathfrak{m}}}{r^{3}},
$$

где $\vec{r}$ — радиус-вектор от диполя до точки наблюдения.

### Диполь во внешнем поле

Во внешнем однородном магнитном поле $\vec{B}$ на диполь действует вращающий **момент сил**:

$$
\vec{M}=[\,\vec{\mathfrak{m}},\,\vec{B}\,]
$$

(векторное произведение). Этот момент стремится развернуть диполь так, чтобы его магнитный момент $\vec{\mathfrak{m}}$ сориентировался вдоль вектора поля $\vec{B}$. Это положение соответствует минимуму потенциальной энергии диполя в поле, которая равна:

$$
W=-(\vec{\mathfrak{m}}\cdot\vec{B}).
$$

В неоднородном поле, помимо вращающего момента, на диполь также действует сила, которая «втягивает» его в область более сильного поля:

$$
\vec{F}=(\vec{\mathfrak{m}}\cdot\nabla)\,\vec{B}.
$$

### Взаимодействие двух диполей

Используя приведённые выше формулы, можно рассчитать силу взаимодействия между двумя магнитами. Для двух небольших магнитов с моментами $\mathfrak{m}_1$ и $\mathfrak{m}_2$, расположенных на расстоянии $r$ друг от друга, сила взаимодействия зависит от их взаимной ориентации.

* Когда моменты направлены **вдоль** линии, соединяющей их центры, сила притяжения (если моменты сонаправлены) равна:

$$
F=\frac{6\,\mathfrak{m}_1\mathfrak{m}_2}{r^{4}}.
$$

* Когда моменты направлены **перпендикулярно** линии, соединяющей их, сила взаимодействия в два раза слабее.

### Магнитное поле Земли

Наша планета создаёт собственное магнитное поле, которое можно в первом приближении представить как поле гигантского магнитного диполя, расположенного в центре Земли. Вектор индукции этого поля $\vec{B}$ в любой точке на поверхности можно разложить на две составляющие:

* **Горизонтальная составляющая** $\bigl(B_{\parallel}\bigr)$, направленная вдоль поверхности Земли к магнитному полюсу.
* **Вертикальная составляющая** $\bigl(B_{\perp}\bigr)$, направленная перпендикулярно поверхности.

Угол $\beta$, который образует полный вектор $\vec{B}$ с горизонтальной плоскостью, называется **магнитным наклонением**:

$$
\beta=\arctan\!\left(\frac{B_{\perp}}{B_{\parallel}}\right).
$$

В данной работе мы измерим обе эти составляющие и определим магнитное наклонение.


## 1. Определение характеристик магнитных шариков

### Оборудование и измерения

Для экспериментов используются неодимовые магниты в форме шариков. С помощью штангенциркуля и весов были определены их базовые параметры:

* Диаметр шарика $(d)$: $0.630 \pm 0.001$ см
* Масса одного шарика $(m)$: $0.82 \pm 0.01$ г

Магнитный момент шариков $\mathfrak{m}$ был определён двумя независимыми методами.

### Метод А: Уравновешивание силой тяжести

Два магнитных шарика удерживают друг друга в поле тяжести Земли на максимальном расстоянии $r_{\max}$, проложив между ними брусок из немагнитного материала. В положении равновесия сила магнитного притяжения уравновешивает силу тяжести, действующую на верхний шарик.



Измеренное максимальное расстояние:

* $r_{\max} = 2.3 \pm 0.4$ см

### Метод Б: Измерение силы сцепления

Составляется длинная цепочка из магнитных шариков, которая подвешивается к одному шарику. Сила, при которой цепочка отрывается, равна силе сцепления, которая учитывает взаимодействие не только с ближайшим соседом, но и со следующими шариками в цепи. Суммарный вклад всех шариков приводит к поправочному коэффициенту $\approx 1.08$.



Минимальная масса цепочки, при которой происходит отрыв:

* $m_{\min} = 350 \pm 1$ г


### Расчёты

```python

import numpy as np

# Константы и измеренные значения
g = 981  # Ускорение свободного падения, см/с^2

# Параметры шариков
d = 0.630  # см
d_err = 0.001 # см
m = 0.82   # г
m_err = 0.01 # г
R = d / 2  # Радиус шарика

# --- Метод А ---
r_max = 2.3  # см
r_max_err = 0.4 # см

# Расчет магнитного момента по методу А
# F_mag = mg => 6 * m_moment^2 / r_max^4 = mg
m_moment_A = np.sqrt(m * g * r_max**4 / 6)

# Расчет погрешности для метода А
m_moment_A_err = m_moment_A * 0.5 * np.sqrt((m_err/m)**2 + (4 * r_max_err / r_max)**2)

print("--- Метод А ---")
print(f"Магнитный момент (m_A): {m_moment_A:.0f} ± {m_moment_A_err:.0f} ед. СГС")

# --- Метод Б ---
m_min = 350 # г
m_min_err = 1 # г
correction_factor = 1.08

# Расчет силы сцепления F0
# F_total = F0 * (1 + 1/2^4 + 1/3^4 + ...) ~ F0 * 1.08
# F_total = m_min * g
F0 = m_min * g / correction_factor
F0_err = m_min_err * g / correction_factor

# Расчет магнитного момента по методу Б
# F0 = 6 * m_moment^2 / d^4
m_moment_B = np.sqrt(F0 * d**4 / 6)

# Расчет погрешности для метода Б
m_moment_B_err = m_moment_B * 0.5 * np.sqrt((F0_err/F0)**2 + (4 * d_err / d)**2)

print("\n--- Метод Б ---")
print(f"Магнитный момент (m_B): {m_moment_B:.0f} ± {m_moment_B_err:.0f} ед. СГС")

# --- Расчет намагниченности и остаточной индукции ---
# Используем результат метода Б, как более точный
V = (4/3) * np.pi * R**3

# Намагниченность M = m_moment / V
M_B = m_moment_B / V
M_B_err = M_B * np.sqrt((m_moment_B_err/m_moment_B)**2 + (3 * d_err / d)**2)

# Остаточная индукция Br = 4 * pi * M
Br_B = 4 * np.pi * M_B
Br_B_err = Br_B * (M_B_err / M_B)

# Индукция на полюсе Bp = 2 * m_moment / R^3
Bp_calc_B = 2 * m_moment_B / R**3
Bp_calc_B_err = Bp_calc_B * np.sqrt((m_moment_B_err/m_moment_B)**2 + (3 * d_err/d)**2)

# Конвертация в мТл (1 Гс = 0.1 мТл)
Br_B_mT = Br_B * 0.1
Br_B_err_mT = Br_B_err * 0.1
Bp_calc_B_mT = Bp_calc_B * 0.1
Bp_calc_B_err_mT = Bp_calc_B_err * 0.1

print("\n--- Характеристики материала (по методу Б) ---")
print(f"Намагниченность (M): {M_B:.0f} ± {M_B_err:.0f} ед. СГС")
print(f"Остаточная индукция (Br): {Br_B_mT:.0f} ± {Br_B_err_mT:.0f} мТл")
print(f"Расчетная индукция на полюсе (Bp): {Bp_calc_B_mT:.0f} ± {Bp_calc_B_err_mT:.0f} мТл")

# Сравнение с прямым измерением
Bp_measured_mT = 260
Bp_measured_err_mT = 5
print(f"Измеренная индукция на полюсе (Bp): {Bp_measured_mT} ± {Bp_measured_err_mT} мТл")
```

### Результаты расчётов

Полученные значения магнитного момента заметно отличаются. Метод Б (по силе сцепления) считается более точным, так как он менее чувствителен к ошибкам в определении расстояния и силам трения. Расчётное значение индукции на полюсе, полученное из метода Б $(360 \pm 22$ мТл$),$ имеет тот же порядок величины, что и измеренное с помощью магнитометра $(260 \pm 5$ мТл$).$ Расхождение может быть связано с неоднородностью намагничивания и погрешностями моделей.

Для дальнейших расчётов будет использоваться значение магнитного момента, полученное **методом Б**: $\mathfrak{m} = 98 \pm 2$ ед. СГС.


## 2. Горизонтальная составляющая магнитного поля Земли $\left(B_{\parallel}\right)$

### Эксперимент

Для измерения $B_{\parallel}$ используется крутильный маятник. «Магнитная стрелка», состоящая из $n$ шариков, подвешивается на тонкой нити в горизонтальном положении. При отклонении от положения равновесия (вдоль магнитного меридиана) возникают крутильные колебания. Период этих колебаний $T$ зависит от момента инерции стрелки $I_n$, её суммарного магнитного момента $n\mathfrak{m}$ и горизонтальной составляющей поля $B_{\parallel}$.

Теоретически, период колебаний $T_n$ прямо пропорционален числу шариков $n$:

$$
T_n=\left(\pi \sqrt{\frac{m d^2}{3\,\mathfrak{m}\,B_{\parallel}}}\right)\cdot n = k \cdot n
$$

где $k$ — коэффициент пропорциональности. Мы измерим зависимость $T(n)$ и из наклона графика определим $B_{\parallel}$.



Были измерены периоды колебаний для стрелок с числом шариков от 4 до 12.

### Данные и график


```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress

# Экспериментальные данные T(n)
n_horiz = np.array([12, 11, 10, 9, 8, 7, 6, 5, 4])
T_horiz = np.array([3.78, 3.53, 3.37, 3.0, 2.73, 2.43, 2.06, 1.75, 1.41]) # секунды
T_err = 0.04  # Погрешность измерения периода, с

# Линейная аппроксимация T(n) = k*n
slope_k, intercept_k, r_value_k, p_value_k, std_err_k = linregress(n_horiz, T_horiz)

# Построение графика
plt.figure(figsize=(10, 6))
plt.errorbar(n_horiz, T_horiz, yerr=T_err, fmt='o', capsize=5, label='Экспериментальные данные')
plt.plot(n_horiz, slope_k * n_horiz + intercept_k, 'r-', label=f'Линейная аппроксимация\n$T = {slope_k:.3f} \\cdot n + {intercept_k:.3f}$')

plt.title('Зависимость периода крутильных колебаний от числа шариков')
plt.xlabel('Число шариков в стрелке, n')
plt.ylabel('Период колебаний, T (с)')
plt.grid(True)
plt.legend()
sns.despine()
plt.show()
```

```python
# Расчет B_||
B_horiz = (np.pi**2 * m * d**2) / (3 * m_moment_B * slope_k**2)

# Расчет погрешности B_||
k_err = std_err_k
B_horiz_err = B_horiz * np.sqrt(
    (m_err/m)**2 + \
    (2*d_err/d)**2 + \
    (m_moment_B_err/m_moment_B)**2 + \
    (2*k_err/slope_k)**2
)

# Вывод результатов в мТл
B_horiz_mT = B_horiz * 0.1
B_horiz_err_mT = B_horiz_err * 0.1

print(f"Наклон графика k = {slope_k:.4f} ± {std_err_k:.4f} с")
print(f"Горизонтальная составляющая B_|| = {B_horiz:.4f} ± {B_horiz_err:.4f} Гс")
print(f"Горизонтальная составляющая B_|| = {B_horiz_mT:.4f} ± {B_horiz_err_mT:.4f} мТл")
```


### Результаты расчётов

График подтверждает линейную зависимость $T(n)$, что говорит о применимости использованной модели. На основе коэффициента наклона прямой была рассчитана горизонтальная составляющая магнитного поля Земли. Было также показано, что влиянием упругости нити можно пренебречь.

**Результат:** $B_{\parallel} = 0.121 \pm 0.007$ Гс $\left(0.0121 \pm 0.0007\ \text{мТл}\right)$.

## 3. Вертикальная составляющая магнитного поля Земли $\left(B_{\perp}\right)$

### Эксперимент

Магнитная стрелка, подвешенная за центр, под действием вертикальной компоненты поля $B_{\perp}$ наклоняется. Чтобы вернуть её в горизонтальное положение, необходимо скомпенсировать магнитный момент сил $M_{\text{mag}}=(n\mathfrak{m})\,B_{\perp}$ моментом силы тяжести от дополнительного груза $M_{\text{grav}}$.

ВСТАВЬТЕ ИЗОБРАЖЕНИЕ УРАВНОВЕШИВАНИЯ СТРЕЛКИ

Измерялся уравновешивающий момент $M$ для стрелок разной длины $n$. Теоретически, эта зависимость также линейна:

$$
M(n) = (\mathfrak{m} B_{\perp}) \cdot n = A \cdot n
$$

Измерив наклон $A$ этой зависимости, можно найти $B_{\perp}$.


### Данные и график

В таблице приведены заранее рассчитанные значения уравновешивающего момента $M$ для разного числа шариков $n$.

```python
import pandas as pd
from IPython.display import display

# Таблица исходных данных M(n)
df_vert = pd.DataFrame({
    "n": [4, 6, 8, 10, 12],
    "M (дин·см)": [146.5, 233.6, 315.2, 526.6, 702.1],
})

# Оценка погрешности для визуализации (если в отчёте не дана — возьмём ~10%)
df_vert["σ_M (дин·см)"] = 0.10 * df_vert["M (дин·см)"]

# Покажем аккуратную таблицу
display(df_vert)

# Достаём массивы для расчётов
n_vert  = df_vert["n"].to_numpy()
M_vert  = df_vert["M (дин·см)"].to_numpy()
M_err   = df_vert["σ_M (дин·см)"].to_numpy()

# Линейная аппроксимация M(n) = A*n
from scipy.stats import linregress
slope_A, intercept_A, r_value_A, p_value_A, std_err_A = linregress(n_vert, M_vert)

# Построение графика
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 5))
plt.errorbar(n_vert, M_vert, yerr=M_err, fmt='o', capsize=5, label='Экспериментальные данные')
plt.plot(n_vert, slope_A * n_vert + intercept_A, label=f'Линейная аппроксимация\n$M = {slope_A:.1f}\\,n + {intercept_A:.1f}$')
plt.title('Зависимость уравновешивающего момента от числа шариков')
plt.xlabel('Число шариков в стрелке, n')
plt.ylabel('Момент сил, M (дин·см)')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# Расчёт B_⊥ c использованием ранее найденного m_moment_B
# A = m_moment_B * B_perp  =>  B_perp = A / m_moment_B
B_vert = slope_A / m_moment_B

# Погрешность B_⊥
A_err = std_err_A
B_vert_err = B_vert * np.sqrt((A_err / slope_A)**2 + (m_moment_B_err / m_moment_B)**2)

# Перевод в мТл (1 Гс = 0.1 мТл)
B_vert_mT = B_vert * 0.1
B_vert_err_mT = B_vert_err * 0.1

print(f"Наклон графика A = {slope_A:.2f} ± {std_err_A:.2f} дин·см на шарик")
print(f"Вертикальная составляющая B_⊥ = {B_vert:.3f} ± {B_vert_err:.3f} Гс")
print(f"Вертикальная составляющая B_⊥ = {B_vert_mT:.3f} ± {B_vert_err_mT:.3f} мТл")
```

### Результаты

График также показывает хорошую линейную зависимость, подтверждая аддитивность магнитных моментов шариков.

**Результат:** $B_{\perp} = 0.72 \pm 0.05$ Гс $\left(0.072 \pm 0.005\ \text{мТл}\right)$.

## 4. Выводы

На основе измеренных компонент можно рассчитать полный вектор индукции магнитного поля и магнитное наклонение в месте проведения эксперимента.

```python
# Расчет полного поля и наклонения
B_total = np.sqrt(B_horiz**2 + B_vert**2)
B_total_err = (1 / B_total) * np.sqrt((B_horiz * B_horiz_err)**2 + (B_vert * B_vert_err)**2)
B_total_mT = B_total * 0.1
B_total_err_mT = B_total_err * 0.1

# Наклонение в градусах
beta = np.arctan(B_vert / B_horiz)
beta_deg = np.degrees(beta)

# Погрешность наклонения
beta_err_deg = np.degrees(
    (1 / (1 + (B_vert/B_horiz)**2)) * \
    np.sqrt((B_vert_err/B_horiz)**2 + (B_vert * B_horiz_err / B_horiz**2)**2)
)

# Теоретическое значение для широты Москвы (phi ~ 56 градусов)
phi_moscow = np.radians(56) # Широта Москвы
beta_theory_deg = np.degrees(np.arctan(2 * np.tan(phi_moscow)))


print("--- Итоговые результаты ---")
print(f"Полный вектор индукции |B| = {B_total:.3f} ± {B_total_err:.3f} Гс")
print(f"Полный вектор индукции |B| = {B_total_mT:.3f} ± {B_total_err_mT:.3f} мТл")
print(f"Магнитное наклонение β = {beta_deg:.1f} ± {beta_err_deg:.1f}°")
print(f"Теоретическое наклонение для широты Москвы ≈ {beta_theory_deg:.1f}°")
```

Полная индукция магнитного поля в лаборатории составила **$\lvert B\rvert = 0.073 \pm 0.005$ мТл**.

Полученные значения, особенно вертикальная компонента, значительно отличаются от типичных значений для Московского региона $\sim 0.05$ мТл. Экспериментальное значение наклонения также сильно превышает теоретическое ($\sim 71^{\circ}$).

Основной причиной таких расхождений, вероятнее всего, является искажение магнитного поля Земли внутри железобетонного здания, а также наличие поблизости электронных приборов и других металлических предметов, что является значительным источником систематической погрешности. Несмотря на это, эксперимент наглядно демонстрирует принципы измерения магнитных полей и свойства постоянных магнитов.
