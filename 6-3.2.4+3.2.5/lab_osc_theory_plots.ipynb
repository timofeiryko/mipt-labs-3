
# Теоретический разбор графиков к лабораторной работе 3.2.4–3.2.5  
**Свободные и вынужденные колебания в электрическом контуре**

Этот ноутбук дополняет основной отчёт и содержит подробное теоретическое обоснование тех зависимостей, которые видны на графиках основного отчёта (график $T^2(C)$, зависимость $1/\Theta^2(1/R^2)$, амплитудно-частотные и фазо-частотные характеристики).


## 1. График зависимости $T^2(C)$ для свободных колебаний

### 1.1. Уравнение движения и собственная частота

Последовательный RLC-контур (катушка $L$, конденсатор $C$, резистор $R$) без внешнего источника описывается уравнением
$$
L \ddot q + R \dot q + \frac{q}{C} = 0,
$$
где $q(t)$ — заряд на конденсаторе.

Разделим уравнение на $L$ и введём обозначения
$$
\gamma = \frac{R}{2L}, \qquad \omega_0^2 = \frac{1}{LC}.
$$
Получаем
$$
\ddot q + 2\gamma \dot q + \omega_0^2 q = 0.
$$

Ищем решение в виде $q(t) = A e^{\lambda t}$. Подстановка даёт характеристическое уравнение
$$
\lambda^2 + 2\gamma \lambda + \omega_0^2 = 0,
$$
откуда
$$
\lambda_{1,2} = -\gamma \pm \sqrt{\gamma^2 - \omega_0^2}.
$$

В случае колебательного режима ($\gamma < \omega_0$) удобно ввести
$$
\omega = \sqrt{\omega_0^2 - \gamma^2},
$$
и решение можно записать в виде
$$
q(t) = q_0 e^{-\gamma t}\cos(\omega t + \varphi_0).
$$

Функции тока и напряжения на конденсаторе имеют тот же вид, например
$$
U_C(t) = U_0 e^{-\gamma t}\cos(\omega t + \varphi_0).
$$

Период затухающих колебаний равен
$$
T = \frac{2\pi}{\omega} = \frac{2\pi}{\sqrt{\omega_0^2 - \gamma^2}}.
$$

### 1.2. Приближение слабого затухания

В нашей установке сопротивление сравнительно невелико, так что $\gamma \ll \omega_0$. В этом случае
$$
\omega = \sqrt{\omega_0^2 - \gamma^2}
      = \omega_0 \sqrt{1 - \frac{\gamma^2}{\omega_0^2}}
      \approx \omega_0 \left(1 - \frac{1}{2}\frac{\gamma^2}{\omega_0^2}\right).
$$

При таком приближении период можно разложить по малому параметру:
$$
T = \frac{2\pi}{\omega}
  \approx \frac{2\pi}{\omega_0}\left(1 + \frac{1}{2}\frac{\gamma^2}{\omega_0^2}\right).
$$

Поправка, связанная с сопротивлением, мала по сравнению с основным членом, поэтому в пределах точности эксперимента достаточно взять
$$
T \approx T_0 = \frac{2\pi}{\omega_0} = 2\pi\sqrt{LC}.
$$

### 1.3. Линейная зависимость $T^2$ от $C$

Возводя приближённое выражение для периода в квадрат, получаем
$$
T^2 \approx (2\pi)^2 LC = 4\pi^2 L\,C.
$$

При фиксированном $L$ это означает, что
$$
T^2 \propto C.
$$

Именно эта линейная зависимость и проверяется на графике $T^2(C)$ основного отчёта: точки экспериментальных значений $T^2$ лежат близко к прямой
$$
T^2 = k C, \qquad k \approx 4\pi^2 L.
$$

По наклону этой прямой можно, в принципе, определить индуктивность $L$ или проверить, насколько реальная катушка близка к заявленному значению.



## 2. График зависимости $1/\Theta^2$ от $1/R^2$ и определение критического сопротивления

На одном из рисунков основного отчёта построен график зависимости $1/\Theta^2$ от $1/R^2$, где $\Theta$ — логарифмический декремент затухания, а $R$ — суммарное активное сопротивление контура. Ниже выводится теоретическая связь между этими величинами.

### 2.1. Логарифмический декремент через параметры контура

Рассмотрим затухающие колебания
$$
U_C(t) = U_0 e^{-\gamma t}\cos(\omega t + \varphi_0).
$$

Максимумы амплитуды повторяются через период $T = 2\pi/\omega$. Пусть $U_k$ и $U_{k+1}$ — амплитуды двух соседних максимумов. Тогда
$$
U_{k+1} = U_k e^{-\gamma T},
$$
и логарифмический декремент определяется как
$$
\Theta = \ln\frac{U_k}{U_{k+1}} = \ln\frac{U_k}{U_k e^{-\gamma T}} = \gamma T.
$$

В режиме слабого затухания удобно выразить период через частоту затухающих колебаний $\omega$:
$$
T = \frac{2\pi}{\omega}.
$$
Тогда
$$
\Theta = \gamma \frac{2\pi}{\omega}.
$$

### 2.2. Добротность и связь с декрементом

Добротность колебательной системы (для второго порядка) часто определяется как
$$
Q = \frac{\omega}{2\gamma},
$$
где $\omega$ — частота затухающего синуса, а $\gamma$ — коэффициент затухания.

Из этого соотношения
$$
\gamma = \frac{\omega}{2Q}.
$$
Подставляя в выражение для $\Theta$,
$$
\Theta = \frac{\omega}{2Q}\cdot \frac{2\pi}{\omega} = \frac{\pi}{Q},
$$
или, что то же самое,
$$
Q = \frac{\pi}{\Theta}.
$$

### 2.3. Добротность как функция сопротивления

Вернёмся к параметрам контура. Для RLC-цепи
$$
\omega_0^2 = \frac{1}{LC}, \qquad \gamma = \frac{R}{2L}.
$$

Частота затухающего синуса равна
$$
\omega = \sqrt{\omega_0^2 - \gamma^2}
       = \sqrt{\frac{1}{LC} - \frac{R^2}{4L^2}}.
$$

Подставляя $\omega$ и $\gamma$ в формулу для добротности,
$$
Q = \frac{\omega}{2\gamma}
  = \frac{1}{2}\sqrt{\frac{\omega_0^2}{\gamma^2} - 1}
  = \frac{1}{2}\sqrt{\frac{\frac{1}{LC}}{\frac{R^2}{4L^2}} - 1}
  = \frac{1}{2}\sqrt{\frac{4L}{C R^2} - 1}.
$$

Получили явную зависимость $Q(R)$ при фиксированных $L$ и $C$:
$$
Q(R) = \frac{1}{2}\sqrt{\frac{4L}{C R^2} - 1}.
$$

### 2.4. Выражение для $\Theta(R)$ и линейность $1/\Theta^2(1/R^2)$

Напомним связь добротности и декремента:
$$
\Theta = \frac{\pi}{Q}.
$$

Подставим найденное выражение для $Q(R)$:
$$
\Theta(R) = \frac{\pi}{\frac{1}{2}\sqrt{\frac{4L}{C R^2} - 1}}
          = \frac{2\pi}{\sqrt{\frac{4L}{C R^2} - 1}}.
$$

Нас интересует величина $1/\Theta^2$:
$$
\frac{1}{\Theta^2}
= \frac{1}{4\pi^2}\left(\frac{4L}{C R^2} - 1\right)
= \frac{L}{\pi^2 C}\cdot\frac{1}{R^2} - \frac{1}{4\pi^2}.
$$

Таким образом, получаем выражение вида
$$
\frac{1}{\Theta^2} = k \cdot \frac{1}{R^2} + b,
$$
где
$$
k = \frac{L}{\pi^2 C}, \qquad b = -\frac{1}{4\pi^2}.
$$

Это означает, что при фиксированных $L$ и $C$ график зависимости $1/\Theta^2$ от $1/R^2$ должен быть прямой линией с наклоном $k$ и отрицательным свободным членом $b$. Именно такая почти линейная зависимость и наблюдается на соответствующем графике основного отчёта.

### 2.5. Критическое сопротивление $R_{\text{кр}}$ из наклона прямой

Критическое сопротивление получается из условия перехода к апериодическому режиму, когда корни характеристического уравнения совпадают и $\omega = 0$:
$$
\omega_0^2 - \gamma^2 = 0
\quad\Rightarrow\quad
\gamma_{\text{кр}} = \omega_0.
$$

С другой стороны, $\gamma = R/(2L)$, поэтому
$$
\frac{R_{\text{кр}}}{2L} = \omega_0 = \frac{1}{\sqrt{LC}},
$$
откуда
$$
R_{\text{кр}} = 2\sqrt{\frac{L}{C}}.
$$

Сравним это выражение со наклоном $k = L/(\pi^2 C)$ найденной прямой. Выразим из него $L/C$:
$$
\frac{L}{C} = k \pi^2.
$$
Тогда
$$
R_{\text{кр}} = 2\sqrt{k \pi^2} = 2\pi\sqrt{k}.
$$

Именно такую формулу для $R_{\text{кр}}$ удобно использовать при обработке эксперимента: по наклону прямой, аппроксимирующей график $1/\Theta^2(1/R^2)$, находят $k$ и затем вычисляют $R_{\text{кр}}$ по формуле
$$
R_{\text{кр}} = 2\pi\sqrt{k}.
$$



## 3. Амплитудно-частотная характеристика (АЧХ) последовательного RLC-контура

На двух графиках основного отчёта изображены экспериментальные АЧХ для двух значений сопротивления $R$. Ниже выводится аналитическая форма зависимости амплитуды напряжения на конденсаторе от частоты внешнего воздействия.

### 3.1. Комплексный импеданс и ток в контуре

Рассмотрим последовательный контур, к которому подключён источник с ЭДС
$$
\mathcal{E}(t) = E_0 \cos(\Omega t).
$$

В комплексной форме напряжение записывается как
$$
\hat{\mathcal{E}}(t) = \hat E_0 e^{i\Omega t},
$$
где $\hat E_0$ — комплексная амплитуда (фазор).

Комплексный импеданс последовательной цепи
$$
Z = R + i\Omega L - \frac{i}{\Omega C}
  = R + i\left(\Omega L - \frac{1}{\Omega C}\right).
$$

Комплексная амплитуда тока равна
$$
\hat I = \frac{\hat E_0}{Z},
$$
а действительная амплитуда тока — это модуль комплексной амплитуды:
$$
I_0 = \frac{E_0}{|Z|}
    = \frac{E_0}{\sqrt{R^2 + \left(\Omega L - \frac{1}{\Omega C}\right)^2}}.
$$

### 3.2. Напряжение на конденсаторе и его амплитуда

Импеданс конденсатора равен
$$
Z_C = -\frac{i}{\Omega C}.
$$

Комплексная амплитуда напряжения на конденсаторе
$$
\hat U_C = \hat I Z_C = \frac{\hat E_0}{Z}\left(-\frac{i}{\Omega C}\right).
$$

Его модуль:
$$
U_{C,0} = |\hat U_C|
        = \frac{E_0}{|Z|}\cdot\frac{1}{\Omega C}
        = \frac{E_0}{\Omega C\sqrt{R^2 + \left(\Omega L - \frac{1}{\Omega C}\right)^2}}.
$$

Это полная частотная зависимость амплитуды напряжения на конденсаторе.

### 3.3. Переход к безразмерной форме через $\omega_0$ и $Q$

Сделаем выражение более наглядным, введя безразмерную частоту
$$
p = \frac{\Omega}{\omega_0},
\qquad
\omega_0 = \frac{1}{\sqrt{LC}},
$$
и добротность
$$
Q = \frac{\omega_0 L}{R} = \frac{1}{R}\sqrt{\frac{L}{C}}.
$$

Сначала упростим реактивный член:
$$
\Omega L - \frac{1}{\Omega C}
= \Omega L - \frac{\omega_0^2 L}{\Omega}
= L\left(\Omega - \frac{\omega_0^2}{\Omega}\right)
= L \Omega \left(1 - \frac{\omega_0^2}{\Omega^2}\right)
= L \Omega \left(1 - \frac{1}{p^2}\right).
$$

Тогда
$$
|Z|^2 = R^2 + \left(\Omega L - \frac{1}{\Omega C}\right)^2
      = R^2 + L^2 \Omega^2\left(1 - \frac{1}{p^2}\right)^2.
$$

Вынесем $R^2$:
$$
|Z|^2 = R^2\left[1 + \left(\frac{L\Omega}{R}\right)^2\left(1 - \frac{1}{p^2}\right)^2\right].
$$

Но
$$
\frac{L\Omega}{R} = \frac{L}{R}\omega_0 p = \frac{Q}{p},
$$
поскольку $Q = \omega_0 L / R$. Тогда
$$
|Z|^2 = R^2\left[1 + \left(\frac{Q}{p}\right)^2\left(1 - \frac{1}{p^2}\right)^2\right].
$$

Теперь рассмотрим множитель перед корнем в выражении для $U_{C,0}$:
$$
\frac{1}{\Omega C} = \frac{1}{p \omega_0 C}.
$$
Но $\omega_0 = 1/\sqrt{LC}$, значит
$$
\omega_0 C = \frac{C}{\sqrt{LC}} = \sqrt{\frac{C}{L}},
$$
поэтому
$$
\frac{1}{\Omega C} = \frac{1}{p}\sqrt{\frac{L}{C}}.
$$

Тогда
$$
U_{C,0} = \frac{E_0}{|Z|}\cdot\frac{1}{\Omega C}
        = \frac{E_0}{R\sqrt{1 + \left(\frac{Q}{p}\right)^2\left(1 - \frac{1}{p^2}\right)^2}}\cdot\frac{1}{p}\sqrt{\frac{L}{C}}.
$$

Перепишем добротность
$$
Q = \frac{1}{R}\sqrt{\frac{L}{C}} \quad\Rightarrow\quad \frac{1}{R}\sqrt{\frac{L}{C}} = Q.
$$
Тогда
$$
U_{C,0} = \frac{E_0 Q}{p\sqrt{1 + \left(\frac{Q}{p}\right)^2\left(1 - \frac{1}{p^2}\right)^2}}.
$$

Можно упростить знаменатель, умножив и разделив на $p$:
$$
U_{C,0} = \frac{E_0 Q}{\sqrt{p^2 + Q^2\left(1 - \frac{1}{p^2}\right)^2}}.
$$

### 3.4. Нормированная АЧХ и резонансная кривая

Нас интересует форма кривой, а не абсолютный масштаб. Введём нормированную амплитуду
$$
A(p) = \frac{U_{C,0}(p)}{U_{C,0}(p_{\text{рез}})}.
$$

Резонанс по напряжению на конденсаторе в последовательном контуре наступает при частоте, очень близкой к $\omega_0$, если добротность достаточно велика. Вблизи резонанса удобно использовать приближение высоких $Q$, при котором выражение для $U_{C,0}$ упрощается и сводится к стандартному виду
$$
A(p) \approx \frac{1}{\sqrt{\left(1 - p^2\right)^2 + \left(\frac{p}{Q}\right)^2}}.
$$

Именно эта функция задаёт форму идеализированной резонансной кривой: при $p = 1$ (на резонансной частоте) амплитуда максимальна, а при уходе от резонанса, как в сторону меньших, так и в сторону больших частот, амплитуда падает.

### 3.5. Связь ширины резонансной кривой и добротности

Ширина резонансного пика на уровне $1/\sqrt{2}$ от максимума (уровень минус 3 дБ) связана с добротностью. Вблизи резонанса можно записать частоту как
$$
\Omega = \omega_0 + \Delta\Omega, \qquad |\Delta\Omega| \ll \omega_0.
$$

При высоких $Q$ получается приближение
$$
A(\Omega) \approx \frac{1}{\sqrt{1 + 4Q^2\left(\frac{\Delta\Omega}{\omega_0}\right)^2}}.
$$

Условие $A = 1/\sqrt{2}$ даёт
$$
1 + 4Q^2\left(\frac{\Delta\Omega}{\omega_0}\right)^2 = 2
\quad\Rightarrow\quad
4Q^2\left(\frac{\Delta\Omega}{\omega_0}\right)^2 = 1
\quad\Rightarrow\quad
|\Delta\Omega| = \frac{\omega_0}{2Q}.
$$

Полная ширина пика на этом уровне равна
$$
2\Delta\Omega = \frac{\omega_0}{Q},
$$
откуда следует классическая связь добротности и ширины резонансной кривой
$$
Q = \frac{\omega_0}{2\Delta\Omega}.
$$

Именно этим соотношением пользуются при определении добротности по экспериментальным АЧХ в основном отчёте: по ширине резонансного пика в координатах частота–амплитуда восстанавливается $Q$.



## 4. Фазо-частотная характеристика (ФЧХ) последовательного RLC-контура

В основном отчёте приведены графики зависимости фазового сдвига между сигналом источника и напряжением на конденсаторе от частоты. Ниже выводится теоретическая форма этой зависимости.

### 4.1. Фаза тока в контуре

Комплексный импеданс цепи
$$
Z = R + i\left(\Omega L - \frac{1}{\Omega C}\right).
$$

Его аргумент (фаза импеданса) равен
$$
\arg Z = \arctan\left(\frac{\Omega L - \frac{1}{\Omega C}}{R}\right).
$$

Комплексная амплитуда тока
$$
\hat I = \frac{\hat E_0}{Z}.
$$

Фаза тока равна
$$
\varphi_I = \arg \hat I = \arg \hat E_0 - \arg Z.
$$

Если считать фазу источника нулевой (то есть $\hat E_0$ вещественной и положительной), то
$$
\varphi_I = -\arg Z
= -\arctan\left(\frac{\Omega L - \frac{1}{\Omega C}}{R}\right).
$$

Таким образом, при низких частотах ток опережает напряжение (конденсатор доминирует), а при высоких — отстаёт (доминирует индуктивность).

### 4.2. Фаза напряжения на конденсаторе

Комплексное напряжение на конденсаторе
$$
\hat U_C = \hat I Z_C, \qquad Z_C = -\frac{i}{\Omega C}.
$$

Фаза напряжения на конденсаторе
$$
\varphi_C = \arg \hat U_C = \arg \hat I + \arg Z_C.
$$

Так как
$$
\arg Z_C = -\frac{\pi}{2},
$$
получаем
$$
\varphi_C = \varphi_I - \frac{\pi}{2}
          = -\arctan\left(\frac{\Omega L - \frac{1}{\Omega C}}{R}\right) - \frac{\pi}{2}.
$$

На практике, при построении ФЧХ, часто отсчитывают фазовый сдвиг между напряжением источника и напряжением на конденсаторе или между током и каким-либо из напряжений. Вид кривой при этом отличается добавлением константы, но форма перехода от одного предельного значения к другому остаётся такой же.

### 4.3. Переход к безразмерной форме через $p = \Omega/\omega_0$ и $Q$

Введём, как и раньше,
$$
p = \frac{\Omega}{\omega_0}, \qquad \omega_0 = \frac{1}{\sqrt{LC}}, \qquad Q = \frac{\omega_0 L}{R}.
$$

Для краткости обозначим
$$
X(\Omega) = \Omega L - \frac{1}{\Omega C}.
$$

Как и в разделе об АЧХ, это выражение можно записать так:
$$
X(\Omega) = L\Omega\left(1 - \frac{1}{p^2}\right).
$$

Тогда отношение $X(\Omega)/R$ выражается через $p$ и $Q$:
$$
\frac{X(\Omega)}{R} = \frac{L\Omega}{R}\left(1 - \frac{1}{p^2}\right)
                   = \frac{Q}{p}\left(1 - \frac{1}{p^2}\right).
$$

Фаза импеданса
$$
\arg Z = \arctan\left(\frac{X(\Omega)}{R}\right)
       = \arctan\left[\frac{Q}{p}\left(1 - \frac{1}{p^2}\right)\right].
$$

Соответственно, фаза тока при нулевой фазе источника
$$
\varphi_I(p) = -\arctan\left[\frac{Q}{p}\left(1 - \frac{1}{p^2}\right)\right].
$$

Фаза напряжения на конденсаторе с учётом $\arg Z_C = -\pi/2$:
$$
\varphi_C(p) = -\arctan\left[\frac{Q}{p}\left(1 - \frac{1}{p^2}\right)\right] - \frac{\pi}{2}.
$$

На практике при обработке ФЧХ удобно работать с разностью фаз $\Delta\varphi$ между двумя сигналами, поэтому графики в отчёте фактически отображают одну из этих функций с точностью до добавления постоянного сдвига.

### 4.4. Характер фазового перехода и оценка добротности

Качественный вид $\varphi(\Omega)$:

- при низких частотах (малые $p$) преобладает ёмкостной характер; фаза тока и напряжений приближается к определённому предельному значению;
- при высоких частотах (большие $p$) преобладает индуктивный характер; фаза стремится к другому пределу;
- вблизи резонанса происходит быстрый переход фазы, тем более резкий, чем больше добротность $Q$.

В линейном приближении около резонансной точки зависимость фазового сдвига от частоты можно представить как
$$
\varphi(\Omega) \approx \frac{\pi}{2} + 2Q\left(\frac{\Omega - \omega_0}{\omega_0}\right),
$$
что позволяет оценивать добротность по диапазону частот, на котором фаза изменяется, например, от $\pi/4$ до $3\pi/4$. Чем уже этот диапазон, тем больше добротность.

Именно таким геометрическим методом (с отражением части кривой относительно горизонтальной прямой и измерением ширины на уровне $\varphi = \pi/4$) пользуются при обработке экспериментальных ФЧХ в основном отчёте.



## 5. Заключение

В этом ноутбуке для каждого типа графиков основного отчёта были выведены аналитические зависимости:

1. Для графика $T^2(C)$ показано, что при слабом затухании период свободных колебаний подчиняется закону
   $$
   T^2 \approx 4\pi^2 L C,
   $$
   так что $T^2$ линейно зависит от ёмкости при фиксированной индуктивности.

2. Для графика $1/\Theta^2(1/R^2)$ получено выражение
   $$
   \frac{1}{\Theta^2} = \frac{L}{\pi^2 C}\cdot\frac{1}{R^2} - \frac{1}{4\pi^2},
   $$
   что объясняет наблюдаемую линейность зависимости и позволяет по наклону прямой восстановить критическое сопротивление $R_{\text{кр}} = 2\sqrt{L/C}$.

3. Для амплитудно-частотной характеристики напряжения на конденсаторе в последовательном контуре получена резонансная кривая вида
   $$
   A(p) \approx \frac{1}{\sqrt{\left(1 - p^2\right)^2 + \left(\frac{p}{Q}\right)^2}},
   $$
   где $p = \Omega/\omega_0$. Показано, что ширина пика на уровне $1/\sqrt{2}$ от максимума связана с добротностью соотношением
   $$
   Q = \frac{\omega_0}{2\Delta\Omega}.
   $$

4. Для фазо-частотной характеристики выведены выражения для фаз тока и напряжения через параметры контура, показана зависимость фазы от безразмерной частоты $p$ и добротности $Q$, а также объяснён характер резкого фазового перехода в области резонанса, используемого для оценки добротности по ФЧХ.

Таким образом, все зависимости, наблюдаемые на графиках основного отчёта, непосредственно следуют из уравнения движения RLC-контура и стандартных определений добротности, логарифмического декремента и комплексного импеданса.
